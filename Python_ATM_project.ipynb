```python
# Initial balance amount
initial_balance = 1000

# Function to display the main menu
def display_main_menu():
    print("\n=== ATM Main Menu ===")
    print("1. Display Bank Account Balance")
    print("2. Withdraw Money")
    print("3. Deposit Money")
    print("4. Exit ATM")

# Function to get PIN from the user and validate it
def get_pin():
    correct_pin = "1234"  # Replace this with your desired PIN
    attempts = 3

    while attempts > 0:
        pin = input("Please enter your 4-digit PIN: ")
        if pin == correct_pin:
            print("PIN accepted. Access granted.")
            return True
        else:
            attempts -= 1
            print(f"Invalid PIN! You have {attempts} attempts left.")

    print("Too many incorrect attempts. Exiting...")
    return False

# Function to display bank account balance
def display_balance(balance):
    print(f"Your current bank account balance is: £{balance:.2f}")

# Function to withdraw money
def withdraw_money(balance):
    print("\n=== Withdrawal Menu ===")
    print("1. £10")
    print("2. £20")
    print("3. £40")
    print("4. £60")
    print("5. Enter other amount")
    print("6. Return to Main Menu")

    choice = input("Please select an option (1/2/3/4/5/6): ")

    if choice == "1":
        return perform_withdrawal(balance, 10)
    elif choice == "2":
        return perform_withdrawal(balance, 20)
    elif choice == "3":
        return perform_withdrawal(balance, 40)
    elif choice == "4":
        return perform_withdrawal(balance, 60)
    elif choice == "5":
        amount = float(input("Enter the amount you want to withdraw (multiple of 10): £"))
        if amount % 10 == 0:
            return perform_withdrawal(balance, amount)
        else:
            print("Withdrawal amount must be a multiple of 10.")
            return balance
    elif choice == "6":
        return balance
    else:
        print("Invalid option. Please choose a valid option.")
        return balance

def perform_withdrawal(balance, amount):
    if amount <= balance:
        balance -= amount
        print(f"Withdrawal of £{amount:.2f} successful.")
    else:
        print("Insufficient funds. Withdrawal failed.")
    return balance

# Function to deposit money
def deposit_money(balance):
    print("\n=== Deposit Menu ===")
    print("1. Deposit funds")
    print("2. Return to Main Menu")

    choice = input("Please select an option (1/2): ")

    if choice == "1":
        amount = float(input("Enter the amount you want to deposit: £"))
        balance += amount
        print(f"Deposit of £{amount:.2f} successful.")
    elif choice == "2":
        return balance
    else:
        print("Invalid option. Please choose a valid option.")

    return balance

# Main program
if get_pin():
    while True:
        display_main_menu()
        choice = input("Please select an option (1/2/3/4): ")

        if choice == "1":
            display_balance(initial_balance)
        elif choice == "2":
            initial_balance = withdraw_money(initial_balance)
        elif choice == "3":
            initial_balance = deposit_money(initial_balance)
        elif choice == "4":
            print("Thank you for using the ATM. Goodbye!")
            break
        else:
            print("Invalid option. Please choose a valid option.")
```
